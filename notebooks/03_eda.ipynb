```python
import pandas as pd
import matplotlib.pyplot as plt

# Display plots in the notebook
%matplotlib inline
```

```python
df = pd.read_csv("../data/cleaned/jobs_cleaned.csv")

df.head()
```

```python
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

df.info()
```

Jobs by Region

```python
region_counts = df["region"].value_counts()

region_counts
```

```python
plt.figure(figsize=(8,5))

region_counts.plot(kind="bar")

plt.title("Jobs by Region")
plt.xlabel("Region")
plt.ylabel("Number of Jobs")

plt.xticks(rotation=45)

plt.show()
```

Top 15 Hiring Companies

```python
top_companies = df["company"].value_counts().head(15)

top_companies
```

```python
plt.figure(figsize=(10,6))

top_companies.sort_values().plot(kind="barh")

plt.title("Top Hiring Companies")
plt.xlabel("Number of Jobs")

plt.show()
```

Job Categories

```python
category_counts = df["category"].value_counts()

category_counts
```

```python
plt.figure(figsize=(7,7))

category_counts.plot(kind="pie", autopct="%1.1f%%")

plt.ylabel("")

plt.title("Job Categories")

plt.show()
```

Top Cities

```python
top_cities = df["city"].value_counts().head(15)

plt.figure(figsize=(10,6))

top_cities.sort_values().plot(kind="barh")

plt.title("Top Cities")

plt.show()
```

Days Open Distribution

```python
plt.figure(figsize=(8,5))

plt.hist(df["days_open"], bins=20)

plt.title("Distribution of Job Advertisement Duration")

plt.xlabel("Days")

plt.ylabel("Jobs")

plt.show()
```

Jobs Posted Per Day

```python
df["posted_date"] = pd.to_datetime(df["posted_date"])

jobs_per_day = df.groupby("posted_date").size()

jobs_per_day
```

```python
plt.figure(figsize=(12,5))

jobs_per_day.plot()

plt.title("Jobs Posted Over Time")

plt.ylabel("Jobs")

plt.show()
```
